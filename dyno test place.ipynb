```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters for the Lorenz attractor
sigma = 10
rho = 28
beta = 8 / 3

# Define the Lorenz vector field
def lorenz_vector_field(x, y, z, sigma, rho, beta):
    dx = sigma * (y - x)
    dy = x * (rho - z) - y
    dz = x * y - beta * z
    return np.array([dx, dy, dz])

# Define a grid in 3D space
x = np.linspace(-20, 20, 10)
y = np.linspace(-30, 30, 10)
z = np.linspace(0, 50, 10)

# Create a meshgrid
X, Y, Z = np.meshgrid(x, y, z)

# Compute the vector field
U, V, W = np.zeros(X.shape), np.zeros(Y.shape), np.zeros(Z.shape)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        for k in range(X.shape[2]):
            vector = lorenz_vector_field(X[i, j, k], Y[i, j, k], Z[i, j, k], sigma, rho, beta)
            U[i, j, k], V[i, j, k], W[i, j, k] = vector

# Plot the vector field
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1.5, normalize=True)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Lorenz Attractor Vector Field")

plt.show()
```
